```python
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

## Relu function

```python
val_relu = np.linspace(-10,10, 100) # these are the sample input values
output_relu = keras.activations.relu(val_relu).numpy() # these are the output array
```

```python
plt.scatter(val_relu, output_relu)
plt.xlabel('input values')
plt.ylabel('output values')
plt.title('Graph for the relu function')
```

## Sigmoid function

```python
val_sig = np.linspace(-10,10,100)
output_sig = keras.activations.sigmoid(val_sig).numpy()
```

```python
plt.scatter(val_sig, output_sig)
plt.xlabel('input values')
plt.ylabel('output values')
plt.title('Graph for the sigmoid function')
```

## tanh function

```python
val_tanh = np.linspace(-10,10,100)
output_tanh = keras.activations.tanh(val_tanh).numpy()
```

```python
plt.scatter(val_tanh, output_tanh)
plt.xlabel('input values')
plt.ylabel('output values')
plt.title('Graph for the tanh function')
```

# Matrices with numpy

```python
X = np.linspace(1,10,10)
Y = np.linspace(10,20,10)

print("The values in X matric")
print(X)

print("The values in Y matric")
print(Y)
```

```python
# addition of two matrices
X + Y
```

```python
# subtraction of the two matices
Y - X
```

```python
# multiplication of two matrices
np.dot(np.linspace(1,5,5), np.linspace(1,5,5))
```

```python
# transpose of the matices 
values_for_trans = np.random.rand(3,2)
transposedArray = np.transpose(values_for_trans)
```

```python
values_for_trans # this is the values with 3X2 matric, before transpose
```

```python
transposedArray # this is the output values with 2X3 matric, after transpose
```
