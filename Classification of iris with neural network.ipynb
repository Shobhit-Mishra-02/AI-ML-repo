```python
# Doing some practice with iris dataset which a classification problem.

import pandas as pd
import numpy as np
```

```python
from sklearn import datasets

data = datasets.load_iris() # laoding the dataset.
```

```python
X = data.data # featrues
Y = data.target # target
```

```python
X
```

```python
Y
```

```python
# Here I need to scale the featrues in the range of 0 to 1 as all the features have got their different value.

from sklearn.preprocessing import MinMaxScaler

minMax = MinMaxScaler()
```

```python
X_scaled = minMax.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, random_state=1, test_size=0.2, stratify=Y)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
from tensorflow import keras
```

```python
# Making a sequential neural network.

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(4,), activation='relu'),
    keras.layers.Dense(6, activation='relu'),
    keras.layers.Dense(4, activation='relu'),
    keras.layers.Dense(3, activation='sigmoid'),
])

model.compile(
    optimizer='adam',
    loss=keras.losses.sparse_categorical_crossentropy,
    metrics=['accuracy'],
    
)
```

```python
model.fit(X_train,Y_train, epochs=1000)
```

```python
acc = model.history.history['accuracy']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(len(acc)), acc)
plt.xlabel("epoches")
plt.ylabel("accuracy")
plt.title('Graph b/w accuracy and epoches')
```

```python
prediction = model.predict(X_test[:5])
```

```python
lt = []

for i in prediction:
    lt.append(np.argmax(i))
    
lt
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
all_prediction = []

for i in model.predict(X_test):
    all_prediction.append(np.argmax(i))
    
all_prediction = np.array(all_prediction)
all_prediction
```

```python
Y_test
```

```python
cfm = confusion_matrix(Y_test, all_prediction)
cfm
```

```python
display = ConfusionMatrixDisplay(cfm)
display.plot()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, all_prediction))
```
